```python
import pandas as pd

# Define the first dictionary containing employee data
data1 = {
    'Name': ['Jai', 'Princi', 'Gaurav', 'Anuj'],
    'Age': [27, 24, 22, 32],
    'Address': ['Nagpur', 'Kanpur', 'Allahabad', 'Kannuaj'],
    'Qualification': ['Msc', 'MA', 'MCA', 'Phd']
}

# Define the second dictionary containing employee data
data2 = {
    'Name': ['Abhi', 'Ayushi', 'Dhiraj', 'Hitesh'],
    'Age': [17, 14, 12, 52],
    'Address': ['Nagpur', 'Kanpur', 'Allahabad', 'Kannuaj'],
    'Qualification': ['Btech', 'B.A', 'Bcom', 'B.hons']
}

# Convert the dictionaries into DataFrames
df = pd.DataFrame(data1, index=[0, 1, 2, 3])
df1 = pd.DataFrame(data2, index=[4, 5, 6, 7])

# Print both DataFrames
print(df, "\n\n", df1)

```

```python
frames=[df,df1]
res1 = pd.concat(frames)
res1
```

```python
import pandas as pd

data1 = {'Name': ['Jai', 'Princi', 'Gaurav', 'Anuj'], 
         'Age': [27, 24, 22, 32], 
         'Address': ['Nagpur', 'Kanpur', 'Allahabad', 'Kannuaj'], 
         'Qualification': ['Msc', 'MA', 'MCA', 'Phd']} 

data2 = {'Name': ['Abhi', 'Ayushi', 'Dhiraj', 'Hitesh'], 
         'Age': [17, 14, 12, 52], 
         'Address': ['Nagpur', 'Kanpur', 'Allahabad', 'Kannuaj'], 
         'Qualification': ['Btech', 'B.A', 'Bcom', 'B.hons']} 

# Convert the dictionaries into DataFrames
df = pd.DataFrame(data1, index=[0, 1, 2, 3])

df1 = pd.DataFrame(data2, index=[4, 5, 6, 7])

print(df, "\n\n", df1)
```

```python
df = pd.DataFrame(data1)
df1 = pd.DataFrame(data2)

print(df , "\n\n" , df1)
res = pd.merge(df,df1,on='Address')
res
```

```python
data1 = {'key':['k0','k1','k2','k3'],
    'Name': ['Jai', 'Princi', 'Gaurav', 'Anuj'], 
         'Age': [27, 24, 22, 32], 
         'Address': ['Nagpur', 'Kanpur', 'Allahabad', 'Kannuaj'], 
         'Qualification': ['Msc', 'MA', 'MCA', 'Phd']} 

data2 = {'key':['k0','k1','k2','k3'],
    'Name': ['Abhi', 'Ayushi', 'Dhiraj', 'Hitesh'], 
         'Age': [17, 14, 12, 52], 
         'Address': ['Nagpur', 'Kanpur', 'Allahabad', 'Kannuaj'], 
         'Qualification': ['Btech', 'B.A', 'Bcom', 'B.hons']} 
df = pd.DataFrame(data1)
df1 = pd.DataFrame(data2)

print(df , "\n\n" , df1)
res = pd.merge(df,df1,on='key')
res
```

```python
data1 = {'Name': ['Jai', 'Princi', 'Gaurav', 'Anuj'], 
         'Age': [27, 24, 22, 32], 
         } 

data2 = {'Address': ['Nagpur', 'Kanpur', 'Allahabad', 'Kannuaj'], 
         'Qualification': ['Btech', 'B.A', 'Bcom', 'B.hons']} 

df = pd.DataFrame(data1,index=['k0','k1','k2','k3'])
df1 = pd.DataFrame(data2,index=['k0','k1','k2','k3'])
print(df , "\n\n" , df1)

res = df.join(df1)
res
```
