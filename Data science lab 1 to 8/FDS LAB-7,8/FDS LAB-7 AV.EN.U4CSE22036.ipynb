7. Visualizing Data by using Histograms, Pie Charts, Heat maps and Box plot

```python
import pandas as pd
data={'Flavor':['Chocolate','Strawberry','Vanilla'],
'Number of Cartons':[16,5,9]
}
icecream=pd.DataFrame(data)
icecream
```

```python
import matplotlib.pyplot as plt
plt.bar(icecream['Flavor'], icecream['Number of Cartons']) 
plt.xlabel('Flavor')
plt.ylabel('Number of Cartons')
plt.title('Number of Ice Cream Cartons by Flavor')  
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.barh(icecream['Flavor'], icecream['Number of Cartons']) 
plt.xlabel('Flavor')
plt.ylabel('Number of Cartons')
plt.title('Number of Ice Cream Cartons by Flavor')  
plt.show()
```

```python
import matplotlib.pyplot as plt
icecream_sorted=icecream.sort_values('Number of Cartons',ascending=True)
plt.barh(icecream['Flavor'], icecream_sorted['Number of Cartons']) 
plt.xlabel('Flavor')
plt.ylabel('Number of Cartons')
plt.title('Number of Ice Cream Cartons by Flavor')  
plt.show()
```
