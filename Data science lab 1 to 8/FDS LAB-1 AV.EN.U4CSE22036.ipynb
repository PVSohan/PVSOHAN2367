1. LIST,DICTIONARY AND TUPLES

```python
num = 7
if num % 2 == 0:
    print("Even")
else:
    print("Odd")
```

```python
n = 10
total = 0
for i in range(1, n + 1):
    total += i
print("Sum:", total)
```

```python
num = 47
is_prime = True
if num > 1:
    for i in range(2, num):
        if num % i == 0:
            is_prime = False
            break
if is_prime:
    print("Prime")
else:
    print("Not Prime")
```

```python
n = 10
a, b = 0, 1
fibonacci_sequence = []
for _ in range(n):
    fibonacci_sequence.append(a)
    a, b = b, a + b
print("Fibonacci:", fibonacci_sequence)
```

```python
my_tuple = (1, 2, 3, 4, 5)
print("First element:", my_tuple[0])
print("Last element:", my_tuple[-1])
```

```python
my_tuple = ('a', 'b', 'c', 'd')
for item in my_tuple:
    print(item)
```
