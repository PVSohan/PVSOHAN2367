```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
actors = pd.read_csv("F:/FOUNDATIONS OF DATA SCIENCE/actors.csv")
```

```python
plt.scatter(actors['Number of Movies'], actors['Total Gross'])
plt.xlabel('Number of Movies')
plt.ylabel('Total Gross')
plt.title('Scatter Plot')
plt.show()
```

```python
plt.scatter(
    actors['Number of Movies'],
    actors['Total Gross'],
    c='pink',          # Color of the markers
    linewidths=2,      # Width of marker edges
    marker='^',        # Shape of the markers (square)
    edgecolor='red'  # Color of the marker edges
)
plt.xlabel('Number of Movies')
plt.ylabel('Total Gross')
plt.title('Scatter Plot')
plt.show()
```

```python
plt.scatter(actors['Number of Movies'],actors['Average per Movie'],c='blue',linewidth=2,marker='o',edgecolor='red',s=50)
plt.xlabel('Number of Movies')
plt.ylabel('Average per Movie')
plt.title('Scatter Plot')
plt.show()
```
