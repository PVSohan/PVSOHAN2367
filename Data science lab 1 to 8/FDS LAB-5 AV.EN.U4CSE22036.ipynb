```python


file_path = 'missing values.xlsx'


excel = pd.read_excel(file_path)

display(excel)
```

```python
import pandas as pd
from sklearn.impute import SimpleImputer



print("Original DataFrame:")
print(excel)

# Identify numerical and categorical 
num_col = excel.select_dtypes(include=['number']).columns
cat_col = excel.select_dtypes(exclude=['number']).columns

# Create imputer objects
imputer_numeric = SimpleImputer(strategy='mean')
imputer_category = SimpleImputer(strategy='most_frequent')

# Apply imputation for numerical columns
excel[num_col] = imputer_numeric.fit_transform(excel[num_col])

# Apply imputation for categorical columns
excel[cat_col] = imputer_category.fit_transform(excel[cat_col])

print("\nDataFrame after imputation:")
print(excel)
```

```python
# Handle missing values
# For numeric columns
numeric_cols = ['Age', 'Fare']
imputer_numeric = SimpleImputer(strategy='median')  # Using median for robustness
df[numeric_cols] = imputer_numeric.fit_transform(df[numeric_cols])

# For categorical columns
categorical_cols = ['Embarked']
imputer_categorical = SimpleImputer(strategy='most_frequent')
df[categorical_cols] = imputer_categorical.fit_transform(df[categorical_cols])

# Drop the 'Cabin' column due to excessive missing values
```
